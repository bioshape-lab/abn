````python
import pickle
````

````python
with open("neural_data.pkl", "rb") as f:
    neural_data = pickle.load(f)

with open("stimulus_metadata.pkl", "rb") as f:
    stimulus_metadata = pickle.load(f)
````

There are **4** human subjects in this dataset, named by integers 1, 2, 5 7

`neural_data` is a nested dictionary:


```python
neural_data[1]['FFA-1']
```
returns a pandas dataframe with 1000 columns (one column per image) and 418 rows (fusiform face area 1 has 418 voxels) for **subject 1**


```python
neural_data[5]['V1v']
```
returns a pandas dataframe with 1000 columns (one column per image) and 457 rows (ventral V1 has 457 voxels) for **subject 5**

`stimulus_metadata` contains information about the 1000 stimulus images (obtained the COCO dataset: https://cocodataset.org/#home)
